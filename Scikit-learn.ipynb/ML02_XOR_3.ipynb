### Keras 모델로 XOR 데이터 정확도 확인해보기

```python
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import accuracy_score

# 1. 데이터(XOR) / Y데이터 변경
x_train = [[0,0], [1,0], [0,1], [1,1]]
y_train = [0, 1, 1, 0]

import numpy as np
x_train = np.array(x_train)  # (4,2)
y_train = np.array(y_train)  # (4, )
# x_train = np.transpose(x_train)
# y_train = np.transpose(y_train)

# 2. 모델링
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(32, input_shape=(2, )))
model.add(Dense(16))
model.add(Dense(1, activation='sigmoid'))

model.summary()

# 3. 훈련
model.compile(loss='mse',optimizer='adam', metrics=['mae'])
model.fit(x_train, y_train, epochs=50, batch_size=1)

# 4. 평가예측
x_test = [[0,0], [1,0], [0,1], [1,1]]
x_test = np.array(x_test)
y_predict = model.predict(x_test)

print(x_test, "의 예측결과 : ", y_predict) 
# 1일 확률을 예측하는 모델
# 만약 [0일확률, 1일 확률] 을 예측하고 싶으면, y(라벨 값)을 원핫인코딩 하고 활성화 함수를 softmax로 변경
```
