```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder

seed = 0
numpy.random.seed(seed)
tf.set_random_seed(seed)


# 1. 붓꽃 데이터 불러오기
dataset = pd.read_csv('./csv/iris.csv', encoding='utf-8',
                      names=['a', 'b', 'c', 'd', 'y'])

# 2. 인풋과 레이블로 나누기
x = dataset.loc[ : , ['a', 'b', 'c', 'd']]
y = dataset.loc[ : , 'y']
x = numpy.array(x)  # (768, 4)
y = numpy.array(y)  # (768, )

# 3. 문자열 인코딩 하기(Label Encoding, One-hot Encoding)
label = LabelEncoder()
y = label.fit_transform(y)  # fit과 transform을 한 줄로 요약하는 방법
#y=label.transform(y)
print(y)

from keras.utils import to_categorical
y = to_categorical(y)
print(y)

# 4. 데이터 나누기
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                    test_size=0.3, 
                                                    random_state=0, 
                                                    shuffle = True)

# 5. 모델 설정(DNN)
model = Sequential()
model.add(Dense(32, input_dim=4, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.summary()

# 6. 모델 컴파일링
model.compile(loss='categorical_crossentropy', # 이진분류는 binary, 다중분류는 categorical
              optimizer='adam',
              metrics=['accuracy'])

# 7. 모델 실행
model.fit(x_train, y_train, epochs=200, batch_size=10)

# 8. 결과 출력
print("\n Accuracy : %.4f" % (model.evaluate(x_test,y_test)[1]))
```
